# Into the black box? Technology sharing in major arms transfers and

beyond

Björn Hagelin  
June 1, 2012

## Summary

Hagelin analyses technology sharing in contemporary major arms transfers, arguing that offsets, industrial participation, and technology transfer have become central bargaining tools for buyers seeking not only advanced weapons but access to otherwise unavailable technologies. Using three linked cases—the multinational F-35 programme, Saab’s Gripen export strategy (including co-development offers), and India’s MMRCA fighter competition plus evolving offset policy—the paper shows how “friendship” is neither necessary nor sufficient for winning major deals; instead, perceived capability fit, supplier reliability, and willingness/ability to share technology dominate. The F-35 case highlights the tension between interoperability benefits of sharing sensitive technology with allies and the US interest in maintaining technology leadership—especially around stealth and software source code (“the black box”). The Indian case illustrates how offsets policy evolves in response to supplier constraints and national industrial ambitions, with India adjusting rules (banking, civil-domain eligibility, and possible ToT-as-offset and multipliers) while trying to accelerate acquisition and build self-sufficiency. The concluding discussion warns of long-run proliferation of manufacturing skills and competitor creation as more buyers demand deep technology access and licence manufacture.

## Key takeaways

-   Offsets/technology transfer have become standard expectations in major arms deals; recipients treat arms transfers as channels for wider technology acquisition.
-   Political friendship helps but does not decide outcomes (e.g., Saab’s losses in Finland/Norway despite proximity and industrial offers).
-   F-35 partnership logic blends interoperability, foreign technology access, cost-sharing, and diplomacy—but implementation generated partner dissatisfaction due to limited guaranteed work share (“best value” rather than juste retour).
-   US technology sharing for F-35 is constrained by disclosure policy and export controls; sensitive areas include stealth and especially **software source code** (“nervous system” / “black box”), which partners appear not to receive even when politically close (UK included).
-   India’s offsets policy evolved (2006 → 2008 → 2011): introduction of banking, widening eligible activities beyond strictly defence to civil aerospace/internal security, and policy debate on ToT eligibility as offsets, multipliers, and improved monitoring/audit.
-   For MMRCA, India raised the effective offset expectation to ~50% (above the 30% baseline) due to programme size; ToT demanded for licensed manufacture is very high (stated as 85% on FX component for the licensed-built portion).
-   Supplier constraints (limited Indian defence absorptive capacity; JV caps; no multipliers) can make high offset demands hard to fulfil; policy revisions partly respond to supplier pushback.
-   Long-run consequences: deep technology transfer and licensed manufacture can allow recipients to become exporters/competitors; skills and production capabilities may proliferate beyond the original seller’s strategic control.

## Realist synthesis coding (CMO)

### CMO 1: Buyer leverage in a “buyer’s market” drives deeper technology demands

-   **Context:** Global arms market characterised as increasingly buyer-favoured; multiple suppliers competing for prestige mega-deals (e.g., India MMRCA).
-   **Mechanism:** Buyers specify high offset shares and extensive ToT/licence manufacture to convert procurement into industrial capability-building.
-   **Outcome:** Suppliers normalise technology sharing as “evil necessity”; offset policies are revised iteratively to maximise feasible extraction of technology/industrial benefits.

### CMO 2: Interoperability requires sharing—yet core “black box” remains protected

-   **Context:** Multinational development/partnering (F-35) aimed at interoperability with allies, cost reduction, and political alignment.
-   **Mechanism:** US shares selected sensitive technologies to support interoperability and partner ROI, but retains control over the most critical enabling technology (source code; key stealth elements) through disclosure policy/export control constraints.
-   **Outcome:** Partners gain participation and some technology/workshare, but remain dependent followers at the leading edge; recurring friction over limited access and restricted tech sharing.

### CMO 3: Small-producer survival strategy depends on co-development + permissive ToT offers

-   **Context:** Smaller producers (Saab) face limited domestic orders and must export; advanced next-gen upgrades require shared development finance.
-   **Mechanism:** Supplier offers generous industrial packages and potential source-code/technology access to overcome non-US status and platform disadvantages.
-   **Outcome:** Some competitiveness gains but no guarantee of success; capability fit and geopolitical weight can override industrial/technology incentives (e.g., Norway choosing F-35).

### CMO 4: Offset-policy liberalisation trades off defence-industrial strengthening vs feasibility

-   **Context:** High defence-only offset shares difficult to meet in limited domestic defence markets; suppliers press for broader eligible offset categories.
-   **Mechanism:** Policy revisions broaden eligible offsets (civil aerospace/internal security), consider ToT as eligible offsets, multipliers, and enhanced monitoring/banking.
-   **Outcome:** Increased supplier feasibility and potential broader economic benefits, but risk of diluting direct strengthening of the domestic military industrial base.
