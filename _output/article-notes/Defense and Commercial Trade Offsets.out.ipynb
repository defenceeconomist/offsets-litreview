# Defense and Commercial Trade Offsets: Impacts on the U.S. Industrial

Base Raise Economic and National Security Concerns

Carol Dawn Petersen  
June 1, 2011

## Summary

Petersen argues that defence and commercial offsets—non-monetary obligations attached to exports of advanced goods—have become a normal condition of international sales, yet US economic policy has remained largely “hands off” in ways that may damage the US industrial base and raise national security risks. Procuring states use offsets to pursue industrial policy aims: technology acquisition, workforce skill-building, employment, regional development, and political acceptability of foreign purchases, using instruments such as co-/licensed production, subcontracting, foreign investment, and “multipliers” that reward favoured activities like technology transfer. The paper contends that firm-level incentives to win contracts can ignore wider national costs: shifting supply chains to foreign suppliers, transferring capabilities and skills abroad, and even helping foreign firms become competitors. Offsets can support allied capability and interoperability, but when they enable advanced technology leakage or hollow out critical industrial skills, they may undermine security. Petersen recommends the Department of Commerce lead a proactive offsets strategy incorporated into national strategy documents.

## Key takeaways

-   Offsets are widely required internationally (the paper notes “over 80 countries” require them) and can reach or exceed 100% of contract value in some years.  
-   Procuring governments use offsets as industrial policy: technology, skills, jobs, sectoral/regional strengthening, and political cover for foreign procurement.  
-   Common offset forms include co-production, licensed production, foreign subcontracting, technology transfer, foreign investment/purchases, and training.  
-   Governments can manage offsets strategically via time limits, penalties, regional distribution requirements, and “banking” of credits for future deals.  
-   “Multipliers” raise credited value of preferred activities (e.g., technology transfer), signalling targeted industrial priorities.  
-   US firms may accept offsets because of global competition and market-entry benefits, even when national costs are externalised.  
-   Risks identified include weakened US suppliers and employment, loss of workforce skills, supply-chain vulnerabilities, and creation of future competitors.  
-   The paper calls for a proactive US offsets strategy led by Commerce (BIS), rather than leaving outcomes to firms and foreign states.

## Realist synthesis coding (CMO)

**CMO 1: Offsets as industrial policy accelerators**  
- **Context:** Procuring governments with explicit industrial, employment, and technology objectives; competitive tendering among exporters.  
- **Mechanism:** Mandated offset obligations (and incentives like multipliers) steer exporter activity into targeted local production/skills/technology transfer.  
- **Outcome:** Faster capability-building in the procuring state’s industrial base and workforce (illustrated with Poland’s F-16-related production and China’s leveraged market access arrangements).

**CMO 2: Market-entry incentives vs national externalities (exporting country)**  
- **Context:** Exporters competing for large contracts and foreign market access; US federal “hands off” posture towards offsets.  
- **Mechanism:** Firms optimise for contract win + cost reductions (new/lower-cost suppliers, joint ventures), discounting impacts on domestic suppliers, skills, and long-run competitiveness.  
- **Outcome:** Potential erosion of the US defence industrial base (supplier displacement, skill loss) and spillovers that reduce net national benefits of export sales.

**CMO 3: Technology transfer and security risk pathway**  
- **Context:** Offsets linked to advanced technological exports; imperfect alignment between export controls and offset-driven production/know-how diffusion.  
- **Mechanism:** Offset fulfilment can expand channels for technology/production capability transfer (including licensed production), increasing chances of diffusion beyond intended end-users.  
- **Outcome:** Elevated national security concerns—especially if leading-edge technologies or production know-how migrate to actors that may later threaten US interests.

**CMO 4: Policy delegation to foreign governments**  
- **Context:** Federal policy treats offsets largely as private transactions; limited proactive strategy despite reporting and interagency dialogue.  
- **Mechanism:** De facto delegation of industrial-base policy to foreign offset requirements and firm fiduciary incentives.  
- **Outcome:** Persistence of offset arrangements that can run counter to US economic and security interests; recommendation for Commerce-led proactive strategy.
