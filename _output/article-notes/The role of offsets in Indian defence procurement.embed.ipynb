# The role of offsets in Indian defense procurement policy

## Summary

Baskaran reviews India’s post-1960s procurement strategy aimed at self-reliance, reduced import dependence, and lower foreign-exchange outflows, arguing that India has relied on offsets in practice even without a formal offsets policy. Three offset options recur throughout Indian procurement: **licensed production/technology transfer**, **countertrade/barter**, and **long-term credit arrangements**. Soviet/Russian deals commonly combined all three, while Western deals more often focused on licensed production/technology assistance plus credit. Outcomes are mixed: licensed production/ToT built notable competence in key Indian defence producers and improved India’s later bargaining power, but core technologies were often withheld and absorption constraints, high import content, delays, and cost overruns meant the technology gap persisted and foreign-exchange savings were limited. Rupee-based countertrade and concessional credit provided short-term foreign-exchange relief, yet contributed to long-run repayment burdens and may have encouraged large-scale procurement. The chapter stresses data limitations and offers a primarily qualitative assessment.

## Key takeaways

-   India systematically used offsets (licensed production/ToT; countertrade/barter; long-term credit) despite lacking a formal offsets framework.
-   From mid-1960s to mid-1980s, deals involved both direct offsets (licensed production/ToT) and indirect offsets (trade in non-defence goods), especially in Soviet/Russian agreements.
-   Early licensed-production programmes (tanks, aircraft, vessels) saw serious absorption problems, delays, cost overruns, and high reliance on imported components/materials.
-   Soviet/Russian ToT was constrained: export restrictions, incomplete technical information, limited post-sale updates, and withholding of “core” technologies.
-   Licensed production nonetheless helped HAL/BEL/MDL accumulate capabilities and strengthened India’s negotiating position in later agreements.
-   Countertrade and long-term credit reduced immediate foreign-exchange pressure, but were associated with hidden burdens (pricing, exchange rate issues, and later debt service) and long-run repayment pressures.
-   Overall, India gained industrial capacity but did not close the technology gap; dependency on advanced direct imports persisted, and foreign-exchange requirements still rose substantially over time.
-   Policy implication: integrate offsets with diversification and indigenous R&D to reduce defence-related debt and foreign-exchange burdens.
