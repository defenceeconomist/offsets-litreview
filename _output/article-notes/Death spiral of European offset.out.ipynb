# Death spiral of European offset?

## Summary

This short article reviews how EU Procurement Directive 2009/81/EC has constrained defence offsets by narrowing the use of Article 346 derogations. It argues that indirect offsets have largely disappeared in Europe and direct offsets are declining, aligning with a broader global trend among advanced economies. The authors contend that many offset policies historically imposed ambitious quotas and penalty structures that delivered limited durable industrial upgrading while embedding cost premia and anti-competitive behaviours. Evidence presented is suggestive rather than comprehensive (Commission reporting, enforcement actions, and contractor survey indications). The paper raises the possibility that offsets persist via extra-Directive channels (collaboration, OCCAR, G2G/FMS), including “under the table” arrangements, and asks whether the Directive accelerates EDTIB consolidation under fiscal constraints and post-Brexit dynamics.

## Key takeaways

-   Directive 2009/81/EC reduced routine reliance on Article 346 to justify offset demands.
-   Indirect offsets effectively eradicated; direct offsets sharply constrained.
-   Decline in offsets mirrors a global shift among advanced economies (e.g., reduced US liabilities; Australia abandoned offsets; Kuwait suspended its programme).
-   Offset regimes often failed to deliver high-value investment/employment; may increase procurement cost (premia, inefficiency, anti-competitive behaviour).
-   Enforcement increased post-transposition: formal notices (2016) and infringement proceedings (2018).
-   Offsets may persist via extra-Directive procurement routes (collaborative programmes, OCCAR, G2G/FMS), including informal/implicit forms.
